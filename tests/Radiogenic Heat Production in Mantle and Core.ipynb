```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import thermal_evolution_layers as th
rad = th.Radiogenics()
```

```python
Hmp = 12e12 # 5.3-10.6 TW from Korenaga 2005, 13-20 TW from Driscoll 2014
Kppm = 200 # 300 ppm from Nimmo 2004
Hcp = rad.ppm2Hp_40K_Nimmo(Kppm) # 1-2 TW from Driscoll 2014.  2 TW = 255 ppm 40K. 

times = np.linspace(0,4.568*rad.CGyr2s, 1000)
Hm = rad.heat_production_mantle(Hmp, times)
Hc = rad.heat_production_core(Hcp, times)

plt.plot(times/rad.CGyr2s, Hm/1e12, label='mantle')
plt.plot(times/rad.CGyr2s, Hc/1e12, label='core')
plt.xlabel('Time (Byr)')
plt.ylabel('Q (TW)')
plt.title('Heat Production through Earth History')
plt.legend(loc=0)
plt.grid()
```
