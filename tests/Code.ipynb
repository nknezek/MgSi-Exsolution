```python
%pylab
%matplotlib inline
import pandas as pd
```

```python
data = pd.read_csv('../computed_solutions_new/run_data.csv')
```

```python
def func_plot(Mg,Si):
    #plt.clf()
    tmp1 = np.where((data['layer_thickness'] == 30.) & (data['X_Mg_0'] == Mg) & (data['X_Si_0'] == Si) & (data['overturn'] == 400))
    plt.scatter(data['T_cmb0'][tmp1[0]],log10(data['nu_present'][tmp1[0]]),s=50,c=data['r_i'][tmp1[0]]/1e3/1250,vmin=0,vmax=2,cmap=plt.cm.RdYlBu)

    del tmp1
    tmp1 = np.where((data['layer_thickness'] == 100.) & (data['X_Mg_0'] == Mg) & (data['X_Si_0'] == Si) & (data['overturn'] == 400))
    plt.scatter(data['T_cmb0'][tmp1[0]],0.25+log10(data['nu_present'][tmp1[0]]),s=50,c=data['r_i'][tmp1[0]]/1e3/1250,vmin=0,vmax=2,cmap=plt.cm.RdYlBu)

    del tmp1
    tmp1 = np.where((data['layer_thickness'] == 1000.) & (data['X_Mg_0'] == Mg) & (data['X_Si_0'] == Si) & (data['overturn'] == 400))
    plt.scatter(data['T_cmb0'][tmp1[0]],0.5+log10(data['nu_present'][tmp1[0]]),s=50,c=data['r_i'][tmp1[0]]/1e3/1250,vmin=0,vmax=2,cmap=plt.cm.RdYlBu)
    plt.title('Mg ={:.3f}; Si = {:.3f}'.format(Mg,Si))
    plt.colorbar()
```

```python
figure(figsize=(14,10))
plt.subplot(331)
func_plot(1e-5,1e-5)
plt.subplot(332)
func_plot(1e-5,0.025)
plt.subplot(333)
func_plot(1e-5,0.05)
plt.subplot(334)
func_plot(0.01,1e-5)
plt.subplot(335)
func_plot(0.01,0.025)
plt.subplot(336)
func_plot(0.01,0.05)
plt.subplot(337)
func_plot(0.05,1e-5)
plt.subplot(338)
func_plot(0.05,0.025)
plt.subplot(339)
func_plot(0.05,0.05)
```
